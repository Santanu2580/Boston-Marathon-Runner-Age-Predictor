# Q4) **Generate the heat map. Explain your observations.**

```python
corr_matrix = df1.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```



1.   First here we create correlation matrix so that the heatmap will show the
correlation values between -1 to 1 for each pair of features.
2.   Now here if any square containing correlation value close to -1 or 1 that means they are highly dependent to each other. That is strongly positive correlation.
3.   Again at the right side of the diagram we can see a vertical straight line bar which show the correlation value and how features are related to each other.

   *   Like if it is dark red then it is strongly positively correlated.
   *   If it is dark blue then it is more negatively correlated.
   *   And from this we can also determine if the value is closed to -0.4 , 0.8 and 0.6 then they are strongly correlated.
   *   And if the value close to 0.0 , 0.2 , -0.2 , 0.4 these are slightly correlated that means these are not much correlated.








# Q5) **Do the exploratory data analysis by generating graphs and plots. Explain your observations.**

```python
print(df1.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df1['age'],kde=True,bins=20,color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```


*   X-axis (Age) represents the different age values within the dataset, ranging from approximately 30 to above 70.
* Y-axis (Frequency) represents the frequency or count of the data points corresponding to each age value.
* The height of each bars shows the number of frequency within a specific age range.
* KDE (Kernel Density Estimate) curve gives the estimate of the probability density function of the variable 'Age'.
* From this curve we can observe that a significant number of individuals in the dataset fall with the range of age 55 to 65.




```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sex',data=df1,palette='pastel',hue="sex")
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```



*   This countplot shows the distribution of the 'sex' feature. Here 0 -> Female and 1 -> Male , and we can see in our data set Male count is more than female count.



```python
plt.figure(figsize=(10, 6))
sns.countplot(x='cp',data=df1,palette='pastel',hue="cp")
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```



*   This plot shows the distribution of different chest pain types.
    * Type 0 :
    This type has the highest count around 140 occurences
    * Type 2 :
    This is the second most common chest pain type around 85 occurences
    * Type 1 :
    This type has a moderate count with approximately 50 occurences
    * Type 3 : This type has the lowest count around 20 occurences


```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df1['chol'])
plt.xticks(np.arange(100,600,50))
plt.show()
```



*   This is the representation of Box Plot which is used for detecting outliers and understanding the distribution of continuous variables across categories.
    * Here the median cholesterol level appears to be just above 250 which suggests that half of the individuals have cholesterol levels below this value and half have levels above it.
    * The IQR range from approximately 200 to 300. This range contains the middle 50% of the cholesterol values in the dataset.
    * The whiskers extend from just below 150 to 350 indicates that most cholesterol values fall within this range.
    * Cholesterol level above 350 can be considered as outliers. These outliers represent individuals with unusually high cholesterol levels compared to the rest of the dataset.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='trestbps', hue='target', data=df1, palette='husl')
plt.title('Age vs. Resting Blood Pressure by Target')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.show()
```



*   A scatter plot is a visual representaion of how two variables are relate to each other.
*   Here if we look at the Scatterplot this show the relationship between 'age' and 'trestbps' with points colored by the 'target' value.
* Here we can observe that the individuals in this dataset seem to be mostly between 40 and 70 years old.
* Resting blood pressure values vary widely across the age groups, with no clear trend but value appear to be concentrated between 120 and 160 for most individuals.
* Again from this scatterplot resting blood pressure above 180 can be considered as outliers.



```python
plt.figure(figsize=(12, 6))
sns.barplot(x='cp', y='thalach', hue='target', data=df1, palette='Set1')
plt.title('Mean Maximum Heart Rate by Chest Pain Type and Target')
plt.xlabel('Chest Pain Type')
plt.ylabel('Mean Maximum Heart Rate')
plt.show()
```



*   This bar plot visualizes the relationship between the mean maximum heart rate, chest pain type, and the target variable.
*   X-axis :
    * Chest Pain Type represents different types of chest pain, which are categorized numerically (0,1,2,3)
* Y-axis :
    * Mean Maximum Heart Rate represents the mean maximum heart rate for individuals.
*   Bars :
    * Red Bars represents the mean maximum rate for individuals with the specified condition like not having heart disease.
    *   Blue Bars represent the mean maximum heart rate for individuals with the specified condition like having heart disease.
* Error Bars :
    * The vertical lines on top of each bar represent the confidence intervals for the mean maximum heart rate. These give an indication of the variability or uncertainty in the esteemed means.

* Chest Pain Type 0 : The mean maximum heart rate is lower for patients with target 0 compared to target 1.
* Chest Pain Type 1, 2 & 3 : The mean maximum heart rate for these types is relatively similar, with patients having target 1 consistently showing a slightly higher or similar mean heart rate compared to those with target 0.
